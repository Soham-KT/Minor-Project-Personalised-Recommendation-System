```python
import pandas as pd
import numpy as np
import polars as pl
```

```python
movies = pd.read_csv('Data/movie.csv')
movies.head()
```

```python
movies.info()
```

```python
movies.describe()
```

```python
movies.isna()
```

```python
movies['genres'] = movies['genres'].apply(lambda x: x.split('|'))
```

```python
movies['genres'].value_counts()
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
```

```python
genres_df = mlb.fit_transform(movies['genres'])
```

```python
print(mlb.classes_)
```

```python
genres_df = pd.DataFrame(genres_df, columns=mlb.classes_)
genres_df.head()
```

```python
genres_df.describe()
```

```python
genres_df[genres_df['(no genres listed)'] == 1].count()
```

```python
movies = pd.concat([movies, genres_df], axis=1)
```

```python
movies.head()
```

```python
movies[movies['Action'] == 1]
```

```python
ratings = pl.read_csv('Data/rating.csv')
ratings.head()
```

```python
ratings.describe()
```

```python
movies.describe()
```

```python
movies_new = pl.from_pandas(movies)
print(movies_new)
```

```python
movies_and_ratings = movies_new.join(ratings, on='movieId')
movies_and_ratings.head()
```

```python
movies_and_ratings = movies_and_ratings.drop(['genres','userId', 'timestamp'])
```

```python
print(movies_and_ratings)
```

```python
new_df = movies_and_ratings.with_columns(
    pl.col(mlb.classes_).cast(pl.Int64)
)
```

```python
print(new_df)
```

```python
movies_and_ratings.write_csv('Data/final_movie_dataset.csv')
```

```python
selected_df = movies_and_ratings.filter(
    pl.any_horizontal(
        (pl.col(['Action', 'Comedy']) == 1)
    )
)
```
